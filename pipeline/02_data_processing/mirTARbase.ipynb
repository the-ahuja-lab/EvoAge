# mirTARbase

```python
import pandas as pd
```

```python
BASE_PATH = "/storage/Arushi/090526_EvoAge/kg_formation/data_collection/"
OUT_PATH  = "/storage/Arushi/090526_EvoAge/kg_formation/processed_data/mirTARbase/"
```

```python
drerio_ncbi = pd.read_csv(
    f'{BASE_PATH}databases_for_mapping/ncbi/Danio_rerio.gene_info',sep = '\t')
drerio_ncbi_id_name_dict = dict(zip(drerio_ncbi['Symbol'], drerio_ncbi['description']))
drerio_ncbi_id_name_dict

cele_ncbi = pd.read_csv(
    f'{BASE_PATH}databases_for_mapping/ncbi/Caenorhabditis_elegans.gene_info',sep = '\t')
cele_ncbi_id_name_dict = dict(zip(cele_ncbi['Symbol'], cele_ncbi['description']))
cele_ncbi_id_name_dict


drosophila_ncbi = pd.read_csv(
    f'{BASE_PATH}databases_for_mapping/ncbi/Drosophila_melanogaster.gene_info',sep = '\t')
drosophila_ncbi_id_name_dict = dict(zip(drosophila_ncbi['Symbol'], drosophila_ncbi['description']))
drosophila_ncbi_id_name_dict


human_ncbi = pd.read_csv(
    f'{BASE_PATH}databases_for_mapping/ncbi/Homo_sapiens.gene_info',sep = '\t')
human_ncbi_id_name_dict = dict(zip(human_ncbi['Symbol'], human_ncbi['description']))
human_ncbi_id_name_dict

mouse_ncbi = pd.read_csv(
    f'{BASE_PATH}databases_for_mapping/ncbi/Mus_musculus.gene_info',sep = '\t')
mouse_ncbi_id_name_dict = dict(zip(mouse_ncbi['Symbol'], mouse_ncbi['description']))
mouse_ncbi_id_name_dict
```

```python
drosophila = pd.read_csv(f'{BASE_PATH}mirTARbase/dme_MTI.csv')
drosophila
```

##  Drosophila file 

```python
drosophila = pd.read_csv(f'{BASE_PATH}mirTARbase/dme_MTI.csv')
drosophila
drosophilaKG = pd.DataFrame({
    'Head': drosophila['miRNA'],
    'Relation': 'Mirna_Gene',
    'Tail': drosophila['Target Gene'],
    'Head_Type': 'miRTarBase_ID',
    'Tail_Type': 'Gene',
    'KG_Source': 'miRTarBase',
    'Head_detail_name': drosophila['miRNA'],
    'Tail_detail_name': '',
    'Species': 'Drosophila melanogaster'
})
drosophilaKG = drosophilaKG[drosophilaKG['Head'] != '0']

drosophilaKG
drosophilaKG['Tail_detail_name'] = drosophilaKG['Tail'].map(drosophila_ncbi_id_name_dict)
drosophilaKG = drosophilaKG[~drosophilaKG['Tail_detail_name'].isna()]
drosophilaKG
```

```python
drosophilaKG.to_csv(f'{OUT_PATH}dmel/miRTarBase_Mirna_Gene.csv',index = None)
del drosophilaKG
```

## Human file 

```python
human = pd.read_csv(f'{BASE_PATH}mirTARbase/hsa_MTI.csv')
human

humanKG = pd.DataFrame({
    'Head': human['miRNA'],
    'Relation': 'Mirna_Gene',
    'Tail': human['Target Gene'],
    'HeadType': 'miRTarBase_ID',
    'TailType': 'Gene',
    'KG_Source': 'miRTarBase',
    'Head_detail_name': human['miRNA'],
    'Tail_detail_name': '',
    'Species': 'Homo sapiens'
})
humanKG = humanKG[humanKG['Head'] != '0']
humanKG
humanKG['Tail_detail_name'] = humanKG['Tail'].map(human_ncbi_id_name_dict)
humanKG = humanKG[~humanKG['Head_detail_name'].isna()]
humanKG = humanKG[~humanKG['Tail_detail_name'].isna()]
humanKG = humanKG[humanKG['Head'].str.startswith('hsa', na=False)]
humanKG
```

```python
humanKG.to_csv(f'{OUT_PATH}hsap/miRTarBase_Mirna_Gene.csv',index = None)
```

```python
del humanKG
```
