# CHEMICAL_PROTEIN

```python
yeast_Chemical_Protein_path = f'{BASE_PATH}stitch/scer/4932.protein_chemical.links.v5.0.tsv'

Y_C_P = pd.read_csv(yeast_Chemical_Protein_path, sep = '\t')

Y_C_P = Y_C_P.rename(columns={
    'chemical': 'head',
    'protein': 'tail'
})


def clean_chemical_id(chem_id):
    return chem_id.lstrip('CIDs').lstrip('CIDm').lstrip('0')

# Apply the function to chemical1 and chemical2 columns only
Y_C_P[['head']] = Y_C_P[['head']].applymap(clean_chemical_id)
Y_C_P

Y_C_P['tail'] = Y_C_P['tail'].str.replace('4932.', '')

Y_C_P['relation'] = 'ChemicalEntity_Gene'
Y_C_P['head_type'] = 'ChemicalEntity'
Y_C_P['tail_type'] = 'Gene'
Y_C_P['relation_type'] = np.nan
Y_C_P['kg_source'] = 'STITCH'

Y_C_P['head_detail_name'] = Y_C_P['head'].astype(str).map(Pubchem_IUPAC_CID_Dict)
Y_C_P['tail_detail_name'] = Y_C_P['tail'].map(NCBI_Yeast_Locustag_Descrip_dict)
Y_C_P['head_id_is'] = 'Pubchem'
Y_C_P['tail_id_is'] = 'SGD_SysematicName'
Y_C_P['head_smiles_name'] = Y_C_P['head'].astype(str).map(Pubchem_CID_Smile_Dict)

Y_C_P['species'] = 'S.cerevisae'
Y_C_P
Y_C_P = Y_C_P[~Y_C_P['head_detail_name'].isna()]
Y_C_P = Y_C_P[~Y_C_P['tail_detail_name'].isna()]
Y_C_P
```

```python
final_df = Y_C_P
```

```python
set(final_df['species']), set(final_df['relation']),set(final_df['head_type']),set(final_df['tail_type']),set(final_df['kg_source']),set(final_df['head_id_is']),set(final_df['tail_id_is'])
```

```python
# Count true NaN values
true_nan_count = final_df.isna().sum()

# Count string 'NAN' values (case-insensitive)
string_nan_count = final_df.apply(lambda col: col.astype(str).str.upper().eq('NAN').sum())

# Combine both counts
nan_summary = pd.DataFrame({
    'NaN_count': true_nan_count,
    "'NAN'_string_count": string_nan_count,
    'Total_NaN_like': true_nan_count + string_nan_count
})
nan_summary
```

```python
# First, concatenate all dataframes as before (assuming `final_df` already exists from earlier step)

# Group by the 3 key columns
group_cols = ['head', 'relation', 'tail']

# For kg_source, combine with '::'
def merge_sources(x):
    return '::'.join(sorted(set(x.dropna())))

# Merge duplicates
final_df_dedup = final_df.groupby(group_cols, as_index=False).agg({
    'head_type': 'first',
    'relation_type': 'first',
    'tail_type': 'first',
    'kg_source': merge_sources,
    'head_id_is': 'first',
    'tail_id_is': 'first',
    'head_detail_name': 'first',
    'tail_detail_name': 'first',
    'species': 'first'
})
final_df_dedup
```
