# CHEMICAL_Chemical

```python
yeast_Chemical_Chemical_path = f'{BASE_PATH}stitch/scer/chemical_chemical.links.v5.0.tsv'

Y_C_C = pd.read_csv(yeast_Chemical_Chemical_path, sep = '\t')

Y_C_C = Y_C_C.rename(columns={
    'chemical1': 'head',
    'chemical2': 'tail'
})

def clean_chemical_id(chem_id):
    return chem_id.lstrip('CIDs').lstrip('CIDm').lstrip('0')

# Apply the function to chemical1 and chemical2 columns only
Y_C_C[['head']] = Y_C_C[['head']].applymap(clean_chemical_id)
Y_C_C[['tail']] = Y_C_C[['tail']].applymap(clean_chemical_id)


Y_C_C['relation'] = 'ChemicalEntity_ChemicalEntity'
Y_C_C['head_type'] = 'ChemicalEntity'
Y_C_C['tail_type'] = 'ChemicalEntity'
Y_C_C['relation_type'] = np.nan

Y_C_C['head_detail_name'] = Y_C_C['head'].map(Pubchem_IUPAC_CID_Dict)
Y_C_C['tail_detail_name'] = Y_C_C['tail'].map(Pubchem_IUPAC_CID_Dict)
Y_C_C['head_smile_name'] = Y_C_C['head'].map(Pubchem_CID_Smile_Dict)
Y_C_C['tail_smile_name'] = Y_C_C['tail'].map(Pubchem_CID_Smile_Dict)
Y_C_C['head_id_is'] = 'Pubchem'
Y_C_C['tail_id_is'] = 'Pubchem'
Y_C_C['species'] = 'S.cerevisae'

Y_C_C
```

```python
Y_C_C['kg_source'] = 'STITCH'
```

```python
final_df = Y_C_C
```

```python
set(final_df['species']), set(final_df['relation']),set(final_df['head_type']),set(final_df['tail_type']),set(final_df['kg_source']),set(final_df['head_id_is']),set(final_df['tail_id_is'])
```

```python
# Count true NaN values
true_nan_count = final_df.isna().sum()

# Count string 'NAN' values (case-insensitive)
string_nan_count = final_df.apply(lambda col: col.astype(str).str.upper().eq('NAN').sum())

# Combine both counts
nan_summary = pd.DataFrame({
    'NaN_count': true_nan_count,
    "'NAN'_string_count": string_nan_count,
    'Total_NaN_like': true_nan_count + string_nan_count
})
nan_summary
```

```python
# First, concatenate all dataframes as before (assuming `final_df` already exists from earlier step)

# Group by the 3 key columns
group_cols = ['head', 'relation', 'tail']

# For kg_source, combine with '::'
def merge_sources(x):
    return '::'.join(sorted(set(x.dropna())))

# Merge duplicates
final_df_dedup = final_df.groupby(group_cols, as_index=False).agg({
    'head_type': 'first',
    'relation_type': 'first',
    'tail_type': 'first',
    'kg_source': merge_sources,
    'head_id_is': 'first',
    'tail_id_is': 'first',
    'head_detail_name': 'first',
    'tail_detail_name': 'first',
    'head_smile_name': 'first',
    'tail_smile_name': 'first',
    'species': 'first'
})
final_df_dedup
```

```python
final_df_dedup.to_csv(f'{OUT_PATH}stitch_YEAST_CHEMICALENTITY_CHEMICALENTITY.csv',index = None)
```
