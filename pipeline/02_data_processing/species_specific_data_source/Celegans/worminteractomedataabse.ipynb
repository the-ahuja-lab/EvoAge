```python
import pandas as pd
import numpy as np
```

```python
# =============================================================================
# BASE PATHS — Update these to match your local directory structure
# =============================================================================
your_path_here = '/storage/Arushi/090526_EvoAge/kg_formation/data_collection/'

# NCBI gene info for C. elegans
NCBI_GENE_INFO_PATH = f'{your_path_here}databases_for_mapping/ncbi/Caenorhabditis_elegans.gene_info'


# Pre-processed WID/FICE gene-gene pickle
WID_PATH = f'{your_path_here}widb/WIDB_integrated_functional_network.txt'

# Final output path
OUTPUT_PATH = '/storage/Arushi/090526_EvoAge/kg_formation/processed_data/wid/wid_CELE_GENE_GENE.csv'
```

```python
# =============================================================================
# Load NCBI gene info and build dictionaries
# =============================================================================

ncbi_cele_gene = pd.read_csv(NCBI_GENE_INFO_PATH, sep='\t')

# Strip 'CELE_' prefix from LocusTag
ncbi_cele_gene['LocusTag'] = ncbi_cele_gene['LocusTag'].str.replace('CELE_', '', regex=False)

# Dictionary: LocusTag → Description (used for head/tail_detail_name)
ncbi_locustag_to_description = dict(zip(ncbi_cele_gene['LocusTag'], ncbi_cele_gene['description']))

print(f"NCBI genes loaded: {len(ncbi_cele_gene):,}")
print(f"LocusTag → Description dictionary size: {len(ncbi_locustag_to_description):,}")
```

```python
C_Ele_G_G_WID = pd.read_csv(f'{your_path_here}/widb/WIDB_integrated_functional_network.txt',sep = '\t')
C_Ele_G_G_WID

C_Ele_G_G_WID = C_Ele_G_G_WID.rename( columns = {'IDA' : 'Head', 'IDB': 'Tail'})

# Drop the specified columns
columns_to_drop = ['WI8','LCI','interologs','genetic','phenotypes','coexpression']
C_Ele_G_G_WID.drop(columns=columns_to_drop, inplace=True)

C_Ele_G_G_WID['Relation'] = 'Gene_Gene'
C_Ele_G_G_WID['Relation_Source'] = 'Worm_Interactome_Database'
C_Ele_G_G_WID['head_type'] = 'Gene'
C_Ele_G_G_WID['tail_type'] = 'Gene'

# Reorder the columns in the DataFrame
desired_columns = ['Head', 'Relation', 'Tail', 'head_type', 'tail_type', 'Relation_Source']
C_Ele_G_G_WID = C_Ele_G_G_WID[desired_columns]
C_Ele_G_G_WID
C_Ele_G_G_WID
```

```python
C_Ele_G_G_WID['relation'] = 'Gene_Gene'
C_Ele_G_G_WID['head_type'] = 'Gene'
C_Ele_G_G_WID['tail_type'] = 'Gene'
C_Ele_G_G_WID['relation_type'] = np.nan
C_Ele_G_G_WID['kg_source'] = C_Ele_G_G_WID['Relation_Source']
C_Ele_G_G_WID['head_detail_name'] = C_Ele_G_G_WID['Head'].map(ncbi_locustag_to_description)
C_Ele_G_G_WID['tail_detail_name'] = C_Ele_G_G_WID['Tail'].map(ncbi_locustag_to_description)
C_Ele_G_G_WID['head_id_is'] = 'NCBI_ID'
C_Ele_G_G_WID['tail_id_is'] = 'NCBI_ID'
C_Ele_G_G_WID['species'] = 'C.elegans'
C_Ele_G_G_WID
C_Ele_G_G_WID
C_Ele_G_G_WID = C_Ele_G_G_WID[~C_Ele_G_G_WID['head_detail_name'].isna()]
C_Ele_G_G_WID = C_Ele_G_G_WID[~C_Ele_G_G_WID['tail_detail_name'].isna()]
C_Ele_G_G_WID
```

```python
C_Ele_G_G_WID.to_csv(f'{OUTPUT_PATH}',index = None)
```
