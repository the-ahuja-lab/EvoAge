# Part 1 — Geroprotective Compounds

Source: `AgeingAtlas/Geroprotective_compounds.csv`  
Relation: `ChemicalEntity_Inhibits/Promotes_BiologicalProcess` → `GO:0007568`

```python
file_gero_comp = pd.read_csv(GERO_COMP_PATH)
print(f"Loaded {len(file_gero_comp):,} rows")
print("Organisms:", set(file_gero_comp['organism']))
file_gero_comp
```

```python
# Standardise organism names and filter to the 5 target species
file_gero_comp.loc[file_gero_comp['organism'] == 'Homo sapiens (cell culture)', 'organism'] = 'Homo sapiens'

Sp_list = ['Caenorhabditis elegans', 'Drosophila melanogaster',
           'Saccharomyces cerevisiae', 'Homo sapiens', 'Mus musculus']
file_gero_comp = file_gero_comp[file_gero_comp['organism'].isin(Sp_list)]

# Normalise phenotype labels → AntiAging / ProAging
anti_aging_terms = ['Extended Healthspan', 'Extended Lifespan ', 'Extended Lifespan /Healthspan',
                    'Senescence  Retardated', 'Senescence  Retardated ']
pro_aging_terms  = ['Shortened Lifespan', 'Shortened lifespan']

file_gero_comp['phenotype'] = file_gero_comp['phenotype'].apply(
    lambda x: 'AntiAging' if x in anti_aging_terms else ('ProAging' if x in pro_aging_terms else x)
)
file_gero_comp['compounds'] = file_gero_comp['compounds'].str.rstrip()

print("Phenotype counts after normalisation:")
print(file_gero_comp['phenotype'].value_counts())
print(f"\nRows remaining: {len(file_gero_comp):,}")
```

```python
# Resolve compound name → PubChem CID (case-insensitive)
Pubchem_Syn_fil_dict_lower = {str(k).lower(): v for k, v in Pubchem_Syn_fil_dict.items()}

file_gero_comp['Head'] = file_gero_comp['compounds'].str.lower().map(Pubchem_Syn_fil_dict_lower)
file_gero_comp['Head'] = file_gero_comp['Head'].astype(str).str.replace(r'\.0$', '', regex=True)

before = len(file_gero_comp)
file_gero_comp = file_gero_comp[file_gero_comp['Head'] != 'nan']
print(f"PubChem CID resolved: {len(file_gero_comp):,} / {before:,} rows kept")
```

```python
# Annotate IUPAC name and SMILES
file_gero_comp['Head_detail_name'] = file_gero_comp['Head'].map(Pubchem_IUPAC_CID_Dict)
file_gero_comp['Head_SMILE_name']  = file_gero_comp['Head'].map(Pubchem_CID_Smile_Dict)
```

```python
# Rename columns, add type/relation/source columns
file_gero_comp = file_gero_comp.rename(columns={
    'compounds': 'Head_Alt_name',
    'phenotype': 'Tail',
    'organism':  'Species'
})

file_gero_comp['Head_type']       = 'ChemicalEntity'
file_gero_comp['Tail_type']       = 'Phenotype'
file_gero_comp['Relation']        = file_gero_comp['Head_type'] + '_' + file_gero_comp['Tail_type']
file_gero_comp['Relation_Source'] = 'Aging_Atlas_gerprotector'
file_gero_comp['Data_type']       = 'Associative'

desired_order = [
    'Head', 'Relation', 'Tail', 'Head_type', 'Tail_type',
    'Head_detail_name', 'Head_SMILE_name',
    'Relation_Source', 'Species', 'Head_Alt_name', 'pmid', 'rnaseq', 'Data_type'
]
existing_cols = [c for c in desired_order if c in file_gero_comp.columns]
file_gero_comp = file_gero_comp[existing_cols]
file_gero_comp
```

```python
# Map phenotype → canonical Relation, set Tail to aging GO term
file_gero_comp['Tail'] = file_gero_comp['Tail'].replace({
    'ProAging':  'ChemicalEntity_Promotes_BiologicalProcess',
    'AntiAging': 'ChemicalEntity_Inhibits_BiologicalProcess',
    'Not Clear': 'ChemicalEntity_NoEffect_BiologicalProcess'
})

# Swap Relation ↔ Tail so Relation column holds the edge type
file_gero_comp[['Relation', 'Tail']] = file_gero_comp[['Tail', 'Relation']]

file_gero_comp['Tail']             = 'GO:0007568'
file_gero_comp['Tail_detail_name'] = 'aging'
file_gero_comp['Tail_type']        = 'BiologicalProcess'
file_gero_comp['Data_type']        = 'Causative'

print("Relation counts:")
print(file_gero_comp['Relation'].value_counts())
print("\nSpecies counts:")
print(file_gero_comp['Species'].value_counts())
```

# Part 2 — Age-related Metabolites

Source: `AgeingAtlas/metabolites_aging_atlas.csv`  
Relation: `ChemicalEntity_Inhibits/Promotes_BiologicalProcess` → `GO:0007568`  
Direction derived from Log2 FC: positive → Pro-Aging, negative → Anti-Aging

```python
file_metabolites = pd.read_csv(METABOLITES_PATH)
file_metabolites
```

```python
file_metabolites = pd.read_csv(METABOLITES_PATH)
print(f"Loaded {len(file_metabolites):,} rows")
print("Species:", file_metabolites['Species'].value_counts().to_dict())
print(f"Log2 FC range: {file_metabolites['Log2 FC'].min():.3f} → {file_metabolites['Log2 FC'].max():.3f}")
file_metabolites = file_metabolites[file_metabolites['Treatment']== 'Young VS Old']
file_metabolites
```

```python
file_metabolites['Treatment'].value_counts()
```

```python
# Resolve metabolite name → PubChem CID
# Fallback to PUBCHEM column if synonym lookup misses
Pubchem_Syn_fil_dict_lower = {str(k).lower(): v for k, v in Pubchem_Syn_fil_dict.items()}

file_metabolites['Head'] = (
    file_metabolites['Biochemical'].str.lower()
    .map(Pubchem_Syn_fil_dict_lower)
    .fillna(file_metabolites['PUBCHEM'])
)
file_metabolites['Head'] = file_metabolites['Head'].astype(str).str.replace(r'\.0$', '', regex=True)

before = len(file_metabolites)
file_metabolites = file_metabolites[file_metabolites['Head'] != 'nan']
print(f"PubChem CID resolved: {len(file_metabolites):,} / {before:,} rows kept")
print("Species after filter:", file_metabolites['Species'].value_counts().to_dict())
```

```python
# Annotate IUPAC name and SMILES
file_metabolites['Head_detail_name'] = file_metabolites['Head'].map(Pubchem_IUPAC_CID_Dict)
file_metabolites['Head_SMILE_name']  = file_metabolites['Head'].map(Pubchem_CID_Smile_Dict)
```

```python
file_metabolites['relation'] = file_metabolites['Log2 FC'].apply(
    lambda x: 'PositivelyAssociatedWithAging' if x < -1   # higher in Old
    else ('NegativelyAssociatedWithAging'     if x > 1   # higher in Young  
    else  'NotAssociatedWithAging')
)
file_metabolites
```

```python
file_metabolites['relation'].value_counts()
```

```python
# Fix GO term
file_metabolites['Tail']             = 'GO:0007568'
file_metabolites['Tail_detail_name'] = 'aging'
file_metabolites['Tail_type']        = 'BiologicalProcess'
file_metabolites['Data_type']        = 'Associative'
file_metabolites
```

```python
# Rename, add type/relation/source columns
file_metabolites = file_metabolites.rename(columns={'Biochemical': 'Head_Alt_name'})

file_metabolites['Head_type']       = 'ChemicalEntity'
file_metabolites['Tail_type']       = 'BiologicalProcess'
file_metabolites['Relation']        = file_metabolites['relation']
file_metabolites['Relation_Source'] = 'Aging_Atlas_Metabolites'
file_metabolites['Data_type']       = 'Associative'

desired_order = [
    'Head', 'Relation', 'Tail', 'Head_type', 'Tail_type',
    'Head_detail_name', 'Head_SMILE_name',
    'Relation_Source', 'Species', 'Head_Alt_name',
    'Cell/Tissue', 'Treatment', 'Log2 FC', 'P'
]
existing_cols = [c for c in desired_order if c in file_metabolites.columns]
file_metabolites = file_metabolites[existing_cols]
file_metabolites
```
