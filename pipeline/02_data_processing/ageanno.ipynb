```python
# ── Load ──────────────────────────────────────────────────────────────────────
Dis_Drug = pd.read_csv(f'{BASE_PATH}ageanno/AgeAnno-main/Disease&Drug/Related_drugs.txt', sep=','
)

# ── Rename columns ────────────────────────────────────────────────────────────
Dis_Drug = Dis_Drug.rename(columns={'geneSymbol': 'Tail', 'ChemicalID': 'Head_MESH'})

print("Unique PubMedIDs:", len(set(Dis_Drug['PubMedIDs'])))

# ── Map MESH → PubChem ID ─────────────────────────────────────────────────────
assert 'ChemicalName' in Dis_Drug.columns, "ChemicalName column missing!"

Dis_Drug['Head'] = (
    Dis_Drug['Head_MESH'].map(Mesh_PUBCHEM_dict)
    .fillna(Dis_Drug['Head_MESH'].map(Mesh_supp_PUBCHEM_dict))
    .fillna(Dis_Drug['ChemicalName'].map(Pubchem_Syn_fil_dict_lower))
)

# Clean float artifacts (e.g. '12345.0' → '12345')
Dis_Drug['Head'] = Dis_Drug['Head'].astype(str).str.replace(r'\.0$', '', regex=True)

# Drop rows where no PubChem ID was found
Dis_Drug = Dis_Drug[Dis_Drug['Head'].notna() & (Dis_Drug['Head'] != 'nan')]

# ── Enrich Head (Chemical) ────────────────────────────────────────────────────
Dis_Drug['Head_detail_name'] = Dis_Drug['Head'].map(Pubchem_IUPAC_CID_Dict)
Dis_Drug['Head_SMILE_name']  = Dis_Drug['Head'].map(Pubchem_CID_Smile_Dict)
```

```python
Dis_Drug = Dis_Drug.rename(columns={'GeneSymbol':'Tail'})
Dis_Drug

# # ── Enrich Tail (Gene) ────────────────────────────────────────────────────────

Dis_Drug['Tail_detail_name'] = Dis_Drug['Tail'].map(NCBI_ALL_GENEname_dict)
Dis_Drug = Dis_Drug[Dis_Drug['Tail_detail_name'] != 'nan']
Dis_Drug
```

```python
Dis_Drug
```

```python

# ── Add metadata columns ──────────────────────────────────────────────────────
Dis_Drug['Head_type']      = 'ChemicalEntity'
Dis_Drug['Tail_type']      = 'Gene'
Dis_Drug['Relation']       = 'ChemicalEntity_Gene'
Dis_Drug['Relation_Source'] = 'AgeAnno'
Dis_Drug['Head_ID_IS']     = 'Pubchem'
Dis_Drug['Tail_ID_IS']     = 'NCBI_ID'
Dis_Drug['Species']        = 'Homo sapiens'

# ── Select & reorder columns ──────────────────────────────────────────────────
desired_cols = [
    "Head", "Relation", "Tail", "Head_type", "Relation_type", "Tail_type",
    "KG_Source", "Head_UBERON_ID", "Head_GO_ID", "Head_HP_ID", "Head_Alias_mapped",
    "Head_detail_name", "Head_ENS", "Head_DO_name", "Head_DO_Alt_ids",
    "Tail_UBERON_ID", "Tail_Alias_mapped", "Tail_detail_name",
    "Head_SMILE_name", "Relation_Source", "Head_ID_IS", "Tail_ID_IS",
    "Tail_GO_ID", "Tail_ENS", "evidence", "InteractionActions", "PubMedIDs", "Species"
]

existing_cols = [col for col in desired_cols if col in Dis_Drug.columns]
Chem_Gene = Dis_Drug[existing_cols]

# ── Save ──────────────────────────────────────────────────────────────────────
Chem_Gene.to_csv(f'{OUT_PATH}ageanno_processed/AgeAnno_Chem_Gene.csv',
    index=False
)
print(f"Saved {len(Chem_Gene)} rows.")
```
