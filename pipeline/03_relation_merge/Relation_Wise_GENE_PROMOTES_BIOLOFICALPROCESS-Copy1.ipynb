```python

agingbank_gene_promotes_BiologicalProcess = pd.read_csv(PROC_DIR + 'agingbank/Human_AgingbankGene_promotes_BioPro.csv')
agingbank_gene_promotes_BiologicalProcess.columns = agingbank_gene_promotes_BiologicalProcess.columns.str.lower()
agingbank_gene_promotes_BiologicalProcess.rename(columns={'Relation_Source': 'kg_source'}, inplace=True)
agingbank_gene_promotes_BiologicalProcess['kg_type'] = 'Aging'
agingbank_gene_promotes_BiologicalProcess['kg_source'] = 'AgingBank'
agingbank_gene_promotes_BiologicalProcess['relation'] = 'Gene_Promotes_BiologicalProcess'

agingbank_gene_promotes_BiologicalProcess['species'] = 'Homo species'
agingbank_gene_promotes_BiologicalProcess
```

```python

agingbank_2_gene_promotes_BiologicalProcess = pd.read_csv(PROC_DIR + 'agingbank/Human_AgingbankGene_promotes_BioPro_file2.csv')
agingbank_2_gene_promotes_BiologicalProcess.columns = agingbank_2_gene_promotes_BiologicalProcess.columns.str.lower()
agingbank_2_gene_promotes_BiologicalProcess.rename(columns={'Relation_Source': 'kg_source'}, inplace=True)
agingbank_2_gene_promotes_BiologicalProcess['kg_type'] = 'Aging'
agingbank_2_gene_promotes_BiologicalProcess['kg_source'] = 'AgingBank'
agingbank_2_gene_promotes_BiologicalProcess['relation'] = 'Gene_Promotes_BiologicalProcess'

agingbank_2_gene_promotes_BiologicalProcess['species'] = 'Homo species'
agingbank_2_gene_promotes_BiologicalProcess
```

```python
SOURCE_DFS = [
    ('CellAge_Human_Gene_Promotes_BioPro_1',    CellAge_Human_Gene_Promotes_BioPro_1),
    ('agingbank_gene_promotes_BiologicalProcess',    agingbank_gene_promotes_BiologicalProcess),
    ('agingbank_2_gene_promotes_BiologicalProcess',    agingbank_2_gene_promotes_BiologicalProcess),
    
]

for name, df in SOURCE_DFS:
    dupes = df.columns[df.columns.duplicated(keep=False)].unique().tolist()
    if dupes:
        print(f"\n[{name}] duplicate columns: {dupes}")
        for col in dupes:
            for i, (_, s) in enumerate(df.loc[:, df.columns == col].items()):
                print(f"  '{col}' occurrence {i} → sample: {s.dropna().head(3).tolist()}")
    else:
        print(f"[{name}] ✓ no duplicates")
```

```python
CONCAT_COLS = [
    'head', 'relation', 'tail',
    'head_type', 'relation_type', 'tail_type',
    'kg_source', 'head_id_is', 'tail_id_is',
    'head_detail_name', 'tail_detail_name','kg_type', 'species'
]

df_list = []
for name, df in SOURCE_DFS:
    tmp = df.loc[:, ~df.columns.duplicated()].copy()
    for col in CONCAT_COLS:
        if col not in tmp.columns:
            tmp[col] = pd.NA
    df_list.append(tmp[CONCAT_COLS])

final_df = pd.concat(df_list, ignore_index=True)
print(f"Combined: {len(final_df):,} rows")
final_df
```

```python
GROUP_COLS = ['head', 'relation', 'tail']

def merge_sources(x):
    return '::'.join(sorted(set(x.dropna())))

final_df_dedup = final_df.groupby(GROUP_COLS, as_index=False).agg({
    'head_type':        'first',
    'relation_type':    'first',
    'tail_type':        'first',
    'kg_source':         merge_sources,
    'head_id_is':       'first',
    'tail_id_is':       'first',
    'head_detail_name': 'first',
    'tail_detail_name': 'first',
    'kg_type': merge_sources,
    'species': 'first',
})
print(f"After deduplication: {len(final_df_dedup):,} rows")
final_df_dedup = final_df_dedup[REQUIRED_COLS]
final_df_dedup.head()
```

```python
def nan_summary(df):
    true_nan   = df.isna().sum()
    string_nan = df.apply(lambda c: c.astype(str).str.upper().eq('NAN').sum())
    return pd.DataFrame({
        'NaN_count':          true_nan,
        "'NAN'_string_count": string_nan,
        'Total_NaN_like':     true_nan + string_nan
    })

nan_summary(final_df_dedup)
```

```python
os.makedirs(os.path.dirname(OUT_PATH), exist_ok=True)
final_df_dedup.to_parquet(OUT_PATH, index=False)
print(f"Saved {len(final_df_dedup):,} rows → {OUT_PATH}")
```
