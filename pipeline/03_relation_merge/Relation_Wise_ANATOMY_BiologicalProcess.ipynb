# ANATOMY ↔ GENE Relation-Wise Merge

Merges ANATOMY–GENE triples from multiple KG sources (DRKG, PrimeKG, Hetionet, TARKG),
aligns to a common schema, deduplicates by `(head, relation, tail)`, and saves the result.

## 0. Configuration

```python
import os
import pandas as pd
import numpy as np

# ── Base directories ──────────────────────────────────────────────────────────
BASE_DIR  = '/storage/Arushi/090526_EvoAge/kg_formation/'
PROC_DIR  = BASE_DIR + 'processed_data/'
PROC_DIR  = BASE_DIR + 'processed_data/'


# ── Output path ───────────────────────────────────────────────────────────────
OUT_PATH  = BASE_DIR + 'processed_data_relation_wise_merge/generalised/ANATOMY_BIOLOGICALPROCESS/ALL_ANATOMY_BIOLOGICALPROCESS.csv'

# ── Required output schema ────────────────────────────────────────────────────
REQUIRED_COLS = [
    'head', 'relation', 'tail',
    'head_type', 'relation_type', 'tail_type',
    'kg_source',
    'head_id_is', 'tail_id_is',
    'head_detail_name', 'tail_detail_name', 'kg_type', 'species'
]
```

## 1. Build Lookup Dictionaries


```python
UBERON = pd.read_csv(f'{BASE_DIR}data_collection/databases_for_mapping/uberon/Uberon_ID_Name_non_dup.csv')
UBERON_dict = dict(zip(UBERON['UBERON ID'], UBERON['Name']))
# UBERON_dict
```

```python
# MESH mapping to map tail (GO_ID) to name
go_db = pd.read_csv(f'{BASE_DIR}data_collection/databases_for_mapping/MESH/MESH_GO_ID_NAME.csv')
GO_dict = dict(zip(go_db['id'], go_db['name']))
del go_db
```

## 2. Load Individual KG Sources

### 2a. Monarch

```python
Monarch_Anatomy_cellularComp = pd.read_csv(f'{BASE_DIR}processed_data/Monarch/Human/Human_AnatomicalEntity_BiologicalProcess_Monarch.csv')
Monarch_Anatomy_cellularComp.columns = Monarch_Anatomy_cellularComp.columns.str.lower()
Monarch_Anatomy_cellularComp['head_detail_name'] = Monarch_Anatomy_cellularComp['head'].map(UBERON_dict)
Monarch_Anatomy_cellularComp['tail_detail_name'] = Monarch_Anatomy_cellularComp['tail'].map(GO_dict)

print(f"DRKG:     {len(Monarch_Anatomy_cellularComp):,} rows | columns: {list(Monarch_Anatomy_cellularComp.columns)}")
Monarch_Anatomy_cellularComp['kg_type'] = 'Generalised'
Monarch_Anatomy_cellularComp['species'] = 'Homo sapiens'
Monarch_Anatomy_cellularComp['kg_source'] = 'Monarch_KG'

Monarch_Anatomy_cellularComp
```

```python
Monarch_Anatomy_cellularComp[Monarch_Anatomy_cellularComp['head_detail_name'].isna()]
```

```python
Monarch_Anatomy_cellularComp[Monarch_Anatomy_cellularComp['tail_detail_name'].isna()]
```

## 3. Align Schemas and Concatenate

```python
source_dfs = [
    Monarch_Anatomy_cellularComp,
]

aligned = []
for df in source_dfs:
    df = df.copy()
    for col in REQUIRED_COLS:
        if col not in df.columns:
            df[col] = None       
    aligned.append(df[REQUIRED_COLS])

final_df = pd.concat(aligned, ignore_index=True)
print(f"Consolidated rows: {len(final_df):,}")
final_df['species'] = 'Homo sapiens'
final_df
```

## 4. QC — NaN Check Before Name Repair

```python
def nan_summary(df):
    true_nan   = df.isna().sum()
    string_nan = df.apply(lambda c: c.astype(str).str.upper().eq('NAN').sum())
    return pd.DataFrame({
        'NaN_count':          true_nan,
        "'NAN'_string_count": string_nan,
        'Total_NaN_like':     true_nan + string_nan
    })

nan_summary(final_df)
```

## 6. Deduplicate and Add Schema Columns

Group by `(head, relation, tail)` and collapse duplicate `kg_source` values with `::` separator.

```python
GROUP_COLS = ['head', 'relation', 'tail']

def merge_sources(x):
    return '::'.join(sorted(set(x.dropna())))

final_df_dedup = final_df.groupby(GROUP_COLS, as_index=False).agg({
    'head_type':        'first',
    'relation_type':    'first',
    'tail_type':        'first',
    'kg_source':        merge_sources,
    'kg_type':          merge_sources,   # ← changed from 'first'
    'head_id_is':       'first',
    'tail_id_is':       'first',
    'head_detail_name': 'first',
    'tail_detail_name': 'first',
    'kg_type': merge_sources,
    'species': 'first',
})

# Enforce final column order
final_df_dedup = final_df_dedup[REQUIRED_COLS]

print(f"After deduplication: {len(final_df_dedup):,} rows")
final_df_dedup.head()
```

## 7. QC — NaN Check After Deduplication

```python
nan_summary(final_df_dedup)
```

```python
final_df_dedup['head_species'] = 'Homo sapiens'
final_df_dedup['tail_species'] = 'Homo sapiens'
```

## 8. Save Output

```python
os.makedirs(os.path.dirname(OUT_PATH), exist_ok=True)
final_df_dedup.to_csv(OUT_PATH, index=False)
print(f"Saved {len(final_df_dedup):,} rows → {OUT_PATH}")
```
