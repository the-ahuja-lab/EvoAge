```python
import os
import pandas as pd
import numpy as np
```

```python
# ── Base directories ──────────────────────────────────────────────────────────
BASE_DIR   = '/storage/Arushi/090526_EvoAge/kg_formation/'
PROC_DIR   = BASE_DIR + 'processed_data/'
DB_DIR     = BASE_DIR + 'data_collection/databases_for_mapping/'

# ── Required output schema ────────────────────────────────────────────────────
REQUIRED_COLS = [
    'head', 'relation', 'tail',
    'head_type', 'relation_type', 'tail_type',
    'kg_source',
    'head_id_is', 'tail_id_is',
    'head_detail_name', 'tail_detail_name', 'kg_type', 'species'
]
```

# POSITIVE ASSOCIATION

```python
# ── Output path ───────────────────────────────────────────────────────────────
OUT_PATH   = BASE_DIR + 'processed_data_relation_wise_merge/generalised/CHEMICAL_POSITIVEASSOC_BIOLOGICALPROCESS/ALL_CHEMICAL_POSITIVEASSOC_BIOLOGICALPROCESS_Human.parquet'
```

```python
# AgeXtend
AgingAtlas_POSS_ASS = pd.read_csv(PROC_DIR + 'agingatlas/Human/ChemicalEntity_PositivelyAssociatedWithAging_BiologicalProcess_Metabolites.csv')
AgingAtlas_POSS_ASS.columns = AgingAtlas_POSS_ASS.columns.str.lower()
# AgingAtlas_POSS_ASS.rename(columns={'relation_source': 'kg_source'}, inplace=True)
AgingAtlas_POSS_ASS['kg_type'] = 'Aging'
AgingAtlas_POSS_ASS['kg_source'] = 'AgingAtlas'
AgingAtlas_POSS_ASS['relation'] = 'ChemicalEntity_PositivelyAssociatedWithAging_BiologicalProcess'
AgingAtlas_POSS_ASS['head_id_is'] = 'Pubchem'
AgingAtlas_POSS_ASS['tail_id_is'] = 'Quick_GO'
AgingAtlas_POSS_ASS['tail_detail_name'] = 'aging'

AgingAtlas_POSS_ASS['species'] = 'Homo sapiens'

AgingAtlas_POSS_ASS
```

```python
SOURCE_DFS = [
    ('AgingAtlas_POSS_ASS', AgingAtlas_POSS_ASS),
]

CONCAT_COLS = [
    'head', 'relation', 'tail',
    'head_type', 'relation_type', 'tail_type',
    'kg_source', 'head_id_is', 'tail_id_is',
    'head_detail_name', 'tail_detail_name', 'kg_type', 'species'
]

df_list = []
for name, df in SOURCE_DFS:
    tmp = df.loc[:, ~df.columns.duplicated()].copy()
    for col in CONCAT_COLS:
        if col not in tmp.columns:
            tmp[col] = pd.NA
    df_list.append(tmp[CONCAT_COLS])

final_df = pd.concat(df_list, ignore_index=True)
print(f'Combined: {len(final_df):,} rows')
final_df

def nan_summary(df):
    true_nan   = df.isna().sum()
    string_nan = df.apply(lambda c: c.astype(str).str.upper().eq('NAN').sum())
    return pd.DataFrame({
        'NaN_count':          true_nan,
        "'NAN'_string_count": string_nan,
        'Total_NaN_like':     true_nan + string_nan
    })

nan_summary(final_df)
```

```python
os.makedirs(os.path.dirname(OUT_PATH), exist_ok=True)
final_df.to_parquet(OUT_PATH, index=False)
print(f'Saved {len(final_df):,} rows → {OUT_PATH}')
```
