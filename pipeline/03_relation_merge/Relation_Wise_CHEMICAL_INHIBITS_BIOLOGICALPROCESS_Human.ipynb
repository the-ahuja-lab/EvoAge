```python
import os
import pandas as pd
import numpy as np
```

```python
# ── Base directories ──────────────────────────────────────────────────────────
BASE_DIR   = '/storage/Arushi/090526_EvoAge/kg_formation/'
PROC_DIR   = BASE_DIR + 'processed_data/'
DB_DIR     = BASE_DIR + 'data_collection/databases_for_mapping/'

# ── Output path ───────────────────────────────────────────────────────────────
OUT_PATH   = BASE_DIR + 'processed_data_relation_wise_merge/generalised/CHEMICAL_INHIBITS_BIOLOGICALPROCESS/ALL_CHEMICAL_INHIBITS_BIOLOGICALPROCESS_Human.parquet'

# ── Required output schema ────────────────────────────────────────────────────
REQUIRED_COLS = [
    'head', 'relation', 'tail',
    'head_type', 'relation_type', 'tail_type',
    'kg_source',
    'head_id_is', 'tail_id_is',
    'head_detail_name', 'tail_detail_name', 'kg_type', 'species'
]
```

```python
# AgeXtend
AgeXtend_Human_Chem_Inhibits_BioPro_1 = pd.read_csv(PROC_DIR + 'agextend_processed/AgeXtend_Human_ChemicalEntity_Inhibits_BiologicalProcess.csv')
AgeXtend_Human_Chem_Inhibits_BioPro_1.columns = AgeXtend_Human_Chem_Inhibits_BioPro_1.columns.str.lower()
AgeXtend_Human_Chem_Inhibits_BioPro_1.rename(columns={'relation_source': 'kg_source'}, inplace=True)
AgeXtend_Human_Chem_Inhibits_BioPro_1['kg_type'] = 'Aging'
AgeXtend_Human_Chem_Inhibits_BioPro_1['kg_source'] = 'AgeXtend'
AgeXtend_Human_Chem_Inhibits_BioPro_1['species'] = 'Homo sapiens'
AgeXtend_Human_Chem_Inhibits_BioPro_1
```

```python
# AgeXtend ALL
AgeXtend_Human_Chem_Inhibits_BioPro_2 = pd.read_csv(PROC_DIR + 'agextend_processed/CHEM_BioP_Hallmarks/All_Chemical_Inhibits_BioPro(Hallmarks).csv')
AgeXtend_Human_Chem_Inhibits_BioPro_2.columns = AgeXtend_Human_Chem_Inhibits_BioPro_2.columns.str.lower()
AgeXtend_Human_Chem_Inhibits_BioPro_2.rename(columns={'relation_source': 'kg_source'}, inplace=True)
AgeXtend_Human_Chem_Inhibits_BioPro_2['kg_type'] = 'Aging'
AgeXtend_Human_Chem_Inhibits_BioPro_2['kg_source'] = 'AgeXtend'
AgeXtend_Human_Chem_Inhibits_BioPro_2['species'] = 'Homo sapiens'
AgeXtend_Human_Chem_Inhibits_BioPro_2
```

```python
# AgingAtlas
AgingAtlas_Human_Chem_Inhibits_BioPro_1 = pd.read_csv(PROC_DIR + 'agingatlas/Human/AgingAtlas_Human_Chemical_Inhibit_BiologicalProcess_Geroprotectors.csv')
AgingAtlas_Human_Chem_Inhibits_BioPro_1.columns = AgingAtlas_Human_Chem_Inhibits_BioPro_1.columns.str.lower()
AgingAtlas_Human_Chem_Inhibits_BioPro_1.rename(columns={'relation_source': 'kg_source'}, inplace=True)
AgingAtlas_Human_Chem_Inhibits_BioPro_1['kg_type'] = 'Aging'
AgingAtlas_Human_Chem_Inhibits_BioPro_1['kg_source'] = 'AgingAtlas'
AgingAtlas_Human_Chem_Inhibits_BioPro_1['species'] = 'Homo sapiens'
AgingAtlas_Human_Chem_Inhibits_BioPro_1
```

```python
SOURCE_DFS = [
    ('AgeXtend_Human_Chem_Inhibits_BioPro_1', AgeXtend_Human_Chem_Inhibits_BioPro_1),
    ('AgeXtend_Human_Chem_Inhibits_BioPro_2', AgeXtend_Human_Chem_Inhibits_BioPro_2),
    ('AgingAtlas_Human_Chem_Inhibits_BioPro_1', AgingAtlas_Human_Chem_Inhibits_BioPro_1),
]

for name, df in SOURCE_DFS:
    dupes = df.columns[df.columns.duplicated(keep=False)].unique().tolist()
    if dupes:
        print(f'\n[{name}] duplicate columns: {dupes}')
        for col in dupes:
            for i, (_, s) in enumerate(df.loc[:, df.columns == col].items()):
                print(f"  '{col}' occurrence {i} → sample: {s.dropna().head(3).tolist()}")
    else:
        print(f'[{name}] ✓ no duplicates')
```

```python
CONCAT_COLS = [
    'head', 'relation', 'tail',
    'head_type', 'relation_type', 'tail_type',
    'kg_source', 'head_id_is', 'tail_id_is',
    'head_detail_name', 'tail_detail_name', 'kg_type', 'species'
]

df_list = []
for name, df in SOURCE_DFS:
    tmp = df.loc[:, ~df.columns.duplicated()].copy()
    for col in CONCAT_COLS:
        if col not in tmp.columns:
            tmp[col] = pd.NA
    df_list.append(tmp[CONCAT_COLS])

final_df = pd.concat(df_list, ignore_index=True)
print(f'Combined: {len(final_df):,} rows')
final_df
```

```python
GROUP_COLS = ['head', 'relation', 'tail']

def merge_sources(x):
    return '::'.join(sorted(set(x.dropna())))

final_df_dedup = final_df.groupby(GROUP_COLS, as_index=False).agg({
    'head_type':        'first',
    'relation_type':    'first',
    'tail_type':        'first',
    'kg_source':         merge_sources,
    'head_id_is':       'first',
    'tail_id_is':       'first',
    'head_detail_name': 'first',
    'tail_detail_name': 'first',
    'kg_type': merge_sources,
    'species': 'first',
})
print(f'After deduplication: {len(final_df_dedup):,} rows')
final_df_dedup = final_df_dedup[REQUIRED_COLS]
final_df_dedup.head()
```

```python
def nan_summary(df):
    true_nan   = df.isna().sum()
    string_nan = df.apply(lambda c: c.astype(str).str.upper().eq('NAN').sum())
    return pd.DataFrame({
        'NaN_count':          true_nan,
        "'NAN'_string_count": string_nan,
        'Total_NaN_like':     true_nan + string_nan
    })

nan_summary(final_df_dedup)
```

```python
final_df_dedup['head_id_is'] = 'Pubchem'
final_df_dedup['tail_id_is'] = 'Quick_GO'
nan_summary(final_df_dedup)
```

```python
os.makedirs(os.path.dirname(OUT_PATH), exist_ok=True)
final_df_dedup.to_parquet(OUT_PATH, index=False)
print(f'Saved {len(final_df_dedup):,} rows → {OUT_PATH}')
```
