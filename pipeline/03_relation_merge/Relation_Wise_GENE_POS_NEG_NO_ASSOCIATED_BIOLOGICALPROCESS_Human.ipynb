```python
import os
import pandas as pd
import numpy as np
```

```python
# ── Base directories ──────────────────────────────────────────────────────────
BASE_DIR   = '/storage/Arushi/090526_EvoAge/kg_formation/'
PROC_DIR   = BASE_DIR + 'processed_data/'
DB_DIR     = BASE_DIR + 'data_collection/databases_for_mapping/'


# ── Required output schema ────────────────────────────────────────────────────
REQUIRED_COLS = [
    'head', 'relation', 'tail',
    'head_type', 'relation_type', 'tail_type',
    'kg_source',
    'head_id_is', 'tail_id_is',
    'head_detail_name', 'tail_detail_name', 'kg_type', 'species'
]
```

```python
# MESH mapping to map tail (GO_ID) to name
go_db = pd.read_csv(DB_DIR + 'MESH/MESH_GO_ID_NAME.csv')
GO_dict = dict(zip(go_db['id'], go_db['name']))
del go_db
```

# POSITIVE ASSOCIATION

```python
# ── Output path ───────────────────────────────────────────────────────────────
OUT_PATH   = BASE_DIR + 'processed_data_relation_wise_merge/generalised/GENE_Positively_BIOLOGICALPROCESS/GENE_Positively_BIOLOGICALPROCESS_Human.parquet'

# DAA
DAA_Human_Gene_Pos_BioPro_1 = pd.read_csv(PROC_DIR + 'digitalagingatlas/Human/DAA_Gene_PositivelyAssociatedWith_BiologicalProcess_Human.csv')
DAA_Human_Gene_Pos_BioPro_1.columns = DAA_Human_Gene_Pos_BioPro_1.columns.str.lower()
DAA_Human_Gene_Pos_BioPro_1['tail_detail_name'] = 'aging'
DAA_Human_Gene_Pos_BioPro_1.rename(columns={'relation_source': 'kg_source'}, inplace=True)
DAA_Human_Gene_Pos_BioPro_1['kg_type'] = 'Aging'
DAA_Human_Gene_Pos_BioPro_1['kg_source'] = 'DAA'
DAA_Human_Gene_Pos_BioPro_1['relation'] = 'Gene_PositivelyAssociatedWith_BiologicalProcess'

DAA_Human_Gene_Pos_BioPro_1['species'] = 'Homo sapiens'
DAA_Human_Gene_Pos_BioPro_1['head_id_is'] = 'NCBI_ID'
DAA_Human_Gene_Pos_BioPro_1['tail_id_is'] = 'Quick_GO'
os.makedirs(os.path.dirname(OUT_PATH), exist_ok=True)
DAA_Human_Gene_Pos_BioPro_1.to_parquet(OUT_PATH, index=False)
print(f'Saved {len(DAA_Human_Gene_Pos_BioPro_1):,} rows → {OUT_PATH}')

DAA_Human_Gene_Pos_BioPro_1
```

```python
def nan_summary(df):
    true_nan   = df.isna().sum()
    string_nan = df.apply(lambda c: c.astype(str).str.upper().eq('NAN').sum())
    return pd.DataFrame({
        'NaN_count':          true_nan,
        "'NAN'_string_count": string_nan,
        'Total_NaN_like':     true_nan + string_nan
    })

nan_summary(DAA_Human_Gene_Pos_BioPro_1)
```
