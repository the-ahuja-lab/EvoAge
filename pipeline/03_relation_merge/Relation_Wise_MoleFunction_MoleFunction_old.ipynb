```python
import pandas as pd

# List of your DataFrame variable names
df_names = [
   'Monarch_MolFxn_MolFxn', 'PrimeKG_Gene_ChemicalEntity1', 'GPKG_MolFxn_MolFxn', 'CrossBAR_MolFxn_MolFxn', 'TARKG_MolFxn_MolFxn', 'TARKG_MolFxn_MolFxn2'
]
# The required columns
required_columns = [
    'head', 'relation', 'tail', 'head_type', 'relation_type', 'tail_type',
    'kg_source', 'head_id_is', 'tail_id_is', 'head_detail_name', 'tail_detail_name'
]

# List to hold cleaned dataframes
df_list = []

# Loop through each DataFrame name, get it from globals(), align columns
for name in df_names:
    df = globals()[name].copy()  # Get the DataFrame object from its name
    for col in required_columns:
        if col not in df.columns:
            df[col] = pd.NA  # Add missing columns with NaN
    df = df[required_columns]  # Reorder/select only required columns
    df_list.append(df)

# Concatenate all into one
final_df = pd.concat(df_list, ignore_index=True)
final_df
```

```python
set(final_df['relation']),set(final_df['head_type']),set(final_df['tail_type']),set(final_df['kg_source']),set(final_df['head_id_is']),set(final_df['tail_id_is'])
```

```python
final_df['head_type'] = 'MolecularFunction'
final_df['tail_type'] = 'MolecularFunction'
final_df['relation'] = 'MolecularFunction_MolecularFunction'
final_df['head_id_is'] = 'Quick_Go'
final_df['tail_id_is'] = 'Quick_Go'
```

```python
set(final_df['relation']),set(final_df['head_type']),set(final_df['tail_type']),set(final_df['kg_source']),set(final_df['head_id_is']),set(final_df['tail_id_is'])
```

```python
set(final_df['relation_type'])
```

```python
# First, concatenate all dataframes as before (assuming `final_df` already exists from earlier step)

# Group by the 3 key columns
group_cols = ['head', 'relation', 'tail']

# For kg_source, combine with '::'
def merge_sources(x):
    return '::'.join(sorted(set(x.dropna())))

# Merge duplicates
final_df_dedup = final_df.groupby(group_cols, as_index=False).agg({
    'head_type': 'first',
    'relation_type': 'first',
    'tail_type': 'first',
    'kg_source': merge_sources,
    'head_id_is': 'first',
    'tail_id_is': 'first',
    'head_detail_name': 'first',
    'tail_detail_name': 'first'
})
final_df_dedup
```

```python
# Count true NaN values
true_nan_count = final_df_dedup.isna().sum()

# Count string 'NAN' values (case-insensitive)
string_nan_count = final_df_dedup.apply(lambda col: col.astype(str).str.upper().eq('NAN').sum())

# Combine both counts
nan_summary = pd.DataFrame({
    'NaN_count': true_nan_count,
    "'NAN'_string_count": string_nan_count,
    'Total_NaN_like': true_nan_count + string_nan_count
})
nan_summary
```

```python
final_df_dedup.to_csv('MOLECULARFUNCTION_MOLECULARFUNCTION/ALL_MOLECULARFUNCTION_MOLECULARFUNCTION.csv',index = None)
```
