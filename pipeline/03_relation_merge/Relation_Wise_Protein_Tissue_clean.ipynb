# Protein ↔ Tissue Relation-Wise Merge


## 0. Configuration

```python
import os
import pandas as pd

BASE_DIR   = '/storage/Arushi/090526_EvoAge/kg_formation/'
PROC_DIR   = BASE_DIR + 'processed_data/'
DB_DIR     = BASE_DIR + 'data_collection/databases_for_mapping/'

OUT_PATH   = BASE_DIR + 'processed_data_relation_wise_merge/generalised/PROTEIN_TISSUE/ALL_PROTEIN_TISSUE.parquet'

REQUIRED_COLS = [
    'head', 'relation', 'tail',
    'head_type', 'relation_type', 'tail_type',
    'kg_source',
    'head_id_is', 'tail_id_is',
    'head_detail_name', 'tail_detail_name', 'kg_type', 'species'
]
```

```python
# RecName dict (AC → RecName)
Uniprot_Recname = pd.read_csv(DB_DIR + 'uniprot/Uniprot_ID_Recname.csv')
Uniprot_Recname_dict = dict(zip(Uniprot_Recname['AC'], Uniprot_Recname['RecName']))

# Parsed TrEMBL dict (AC → NAME) for head/tail name fallback
uniprot = pd.read_csv(DB_DIR + 'uniprot/uniprot_parsed_results.csv')
uniprot_trembl_AC_Name_dict = dict(zip(uniprot['AC'], uniprot['NAME']))
del uniprot

print(f"UniProt RecName dict: {len(Uniprot_Recname_dict):,} | TrEMBL dict: {len(uniprot_trembl_AC_Name_dict):,}")
```

## 1. Load KG Sources

### 1a. ageanno_mo

```python
ageanno_mo = pd.read_csv(PROC_DIR + 'ageanno_processed_mo/AgeAnnoMO_Human_Protein_Tissue.csv')
ageanno_mo.rename(columns={'relation_source': 'kg_source', 'tail_name': 'head_detail_name'}, inplace=True)
ageanno_mo.columns = ageanno_mo.columns.str.lower()
print(f"ageanno_mo: {len(ageanno_mo):,} rows | columns: {list(ageanno_mo.columns)}")

ageanno_mo['kg_type'] = 'Aging'
ageanno_mo['kg_source'] = 'AgeAnnoMO'
ageanno_mo['species'] = 'Homo species'
ageanno_mo.head(2)
```

### 1b. ckg

```python
ckg = pd.read_csv(PROC_DIR + 'CKG/File_7_ProtQC_bioM_Tissue_CKG.csv')
ckg.rename(columns={'relation_source': 'kg_source', 'tail_name': 'head_detail_name'}, inplace=True)
ckg.columns = ckg.columns.str.lower()
print(f"ckg: {len(ckg):,} rows | columns: {list(ckg.columns)}")
ckg['kg_type'] = 'Generalised'
ckg['kg_source'] = 'CKG'
ckg['species'] = 'Homo species'
ckg.head(2)
```

```python
ckg
```

```python
ckg_2 = pd.read_csv(PROC_DIR + 'CKG/File_23_protein_tissue_CKG.csv')
ckg_2.rename(columns={'relation_source': 'kg_source', 'tail_name': 'head_detail_name'}, inplace=True)
ckg_2.columns = ckg_2.columns.str.lower()
print(f"ckg_2: {len(ckg_2):,} rows | columns: {list(ckg_2.columns)}")
ckg_2.rename(columns={'head_alt_name':'head_detail_name','tail_bto_name':'tail_detail_name'}, inplace=True, errors='ignore')
ckg_2['kg_type'] = 'Generalised'
ckg_2['kg_source'] = 'CKG'
ckg_2['species'] = 'Homo species'
ckg_2
```

## 2. Check and Fix Duplicate Columns

```python
SOURCE_DFS = [('ageanno_mo', ageanno_mo), # aging
              ('ckg', ckg),
              ('ckg_2', ckg_2)
             ]

for name, df in SOURCE_DFS:
    dupes = df.columns[df.columns.duplicated(keep=False)].unique().tolist()
    print(f"[{name}] {'duplicate columns: ' + str(dupes) if dupes else '✓ no duplicates'}")
```

## 3. Align Schemas and Concatenate

```python
CONCAT_COLS = [
    'head', 'relation', 'tail',
    'head_type', 'relation_type', 'tail_type',
    'kg_source', 'head_id_is', 'tail_id_is',
    'head_detail_name', 'tail_detail_name','kg_type', 'species'
]

df_list = []
for name, df in SOURCE_DFS:
    tmp = df.loc[:, ~df.columns.duplicated()].copy()
    for col in CONCAT_COLS:
        if col not in tmp.columns:
            tmp[col] = pd.NA
    df_list.append(tmp[CONCAT_COLS])

final_df = pd.concat(df_list, ignore_index=True)
print(f"Combined: {len(final_df):,} rows")
final_df.head(2)
```

```python
final_df
```

## 4. Standardise Fixed-Value Columns

```python
final_df['head_type'] = 'Protein'
final_df['tail_type'] = 'Tissue'
final_df['relation']  = 'Protein_Tissue'
final_df['head_id_is']  = 'Uniprot_ID'
final_df['tail_id_is']  = 'BTO_ID'

print("Unique relation:",   set(final_df['relation']))
print("Unique head_id_is:", set(final_df['head_id_is']))
print("Unique tail_id_is:", set(final_df['tail_id_is']))
print("Unique kg_source:",  set(final_df['kg_source']))
print("Unique kg_source:",  set(final_df['kg_type']))
```

```python
final_df[final_df['head_detail_name'].isna()]
```

```python
final_df[final_df['tail_detail_name'].isna()]
```

## 5. Deduplicate and Add Schema Columns

```python
GROUP_COLS = ['head', 'relation', 'tail']

def merge_sources(x):
    return '::'.join(sorted(set(x.dropna())))

final_df_dedup = final_df.groupby(GROUP_COLS, as_index=False).agg({
    'head_type':        'first',
    'relation_type':    'first',
    'tail_type':        'first',
    'kg_source':         merge_sources,
    'head_id_is':       'first',
    'tail_id_is':       'first',
    'head_detail_name': 'first',
    'tail_detail_name': 'first',
    'kg_type': merge_sources,
    'species': 'first',
})
print(f"After deduplication: {len(final_df_dedup):,} rows")
final_df_dedup = final_df_dedup[REQUIRED_COLS]
final_df_dedup.head()
```

## 6. QC — NaN Check

```python
def nan_summary(df):
    true_nan   = df.isna().sum()
    string_nan = df.apply(lambda c: c.astype(str).str.upper().eq('NAN').sum())
    return pd.DataFrame({
        'NaN_count':          true_nan,
        "'NAN'_string_count": string_nan,
        'Total_NaN_like':     true_nan + string_nan
    })

nan_summary(final_df_dedup)
```

```python
final_df_dedup['head_detail_name'] = (
    final_df_dedup['head_detail_name']
    .fillna(
        final_df_dedup['head'].map(uniprot_trembl_AC_Name_dict)
        .fillna(
            final_df_dedup['head'].map(Uniprot_Recname_dict)
        )
    )
)
```

```python
nan_summary(final_df_dedup)
```

```python
final_df_dedup = final_df_dedup[~final_df_dedup['tail_detail_name'].isna()]
final_df_dedup = final_df_dedup[~final_df_dedup['head_detail_name'].isna()]
final_df_dedup
```

## 7. Save Output

```python
os.makedirs(os.path.dirname(OUT_PATH), exist_ok=True)
final_df_dedup.to_parquet(OUT_PATH, index=False)
print(f"Saved {len(final_df_dedup):,} rows → {OUT_PATH}")
```
