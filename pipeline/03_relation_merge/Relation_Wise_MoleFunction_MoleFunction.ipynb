# MolecularFunction → MolecularFunction Relation Pipeline

Builds a unified, deduplicated edge table for the **MolecularFunction–MolecularFunction** relation.

**Output schema:** `head | relation | tail | head_type | relation_type | tail_type | kg_source | kg_type | head_id_is | tail_id_is | head_detail_name | tail_detail_name`

---
**Sources included:**
- Monarch (Human_MolecularActivity_MolecularActivity_Monarch)\n- PrimeKG (PrimekG_MolFxn_MolFxn)\n- GPKG (MolecularFunction_MolecularFunction_GPKG)\n- Crossbar (MolecularFunction_MolecularFunction_Crossbar)\n- TARKG (MolecularFunction_MolecularFunction)\n- TARKG (MolecularFunction_MolecularFunction_2)

**Base paths:**
- KG data: `/storage/Arushi/090526_EvoAge/kg_formation/`
- Processed KG files: `/storage/Arushi/090526_EvoAge/kg_formation/processed_data/`

## 0 · Imports & Base Paths

```python
import pandas as pd
import os

BASE_DIR     = '/storage/Arushi/090526_EvoAge/kg_formation/'
PROC_DIR     = BASE_DIR + 'processed_data/'
MAPPING_DIR  = BASE_DIR + 'data_collection/databases_for_mapping/'
OUT_PATH = BASE_DIR + 'processed_data_relation_wise_merge/generalised/MOLECULARFUNCTION_MOLECULARFUNCTION/ALL_MOLECULARFUNCTION_MOLECULARFUNCTION.csv'

REQUIRED_COLS = [
    'head', 'relation', 'tail', 'head_type', 'relation_type', 'tail_type',
    'kg_source', 'kg_type', 'head_id_is', 'tail_id_is', 'head_detail_name', 'tail_detail_name'
]
```

## 1 · Mapping DBs

```python
# GO Dictionary mapping
GO = pd.read_csv(MAPPING_DIR + 'MESH/MESH_GO_ID_NAME.csv')
GO_dict = dict(zip(GO['id'], GO['name']))
del GO
```

## 2 · Load Source Files

```python
# 1. Monarch (MolecularActivity -> MolecularActivity: Native)
Monarch = pd.read_csv(PROC_DIR + 'Monarch/Human/Human_MolecularActivity_MolecularActivity_Monarch.csv')
Monarch.columns = Monarch.columns.str.lower()
if 'kgsource' in Monarch.columns:
    Monarch = Monarch.rename(columns={'kgsource': 'kg_source'})
if 'head_name' in Monarch.columns:
    Monarch = Monarch.rename(columns={'head_name': 'head_detail_name'})
if 'tail_name' in Monarch.columns:
    Monarch = Monarch.rename(columns={'tail_name': 'tail_detail_name'})


required_columns = [
    'head', 'relation', 'tail', 'head_type', 'relation_type', 'tail_type',
    'kg_source', 'head_id_is', 'tail_id_is', 'head_detail_name', 'tail_detail_name'
]
Monarch = Monarch[required_columns]
Monarch['head_id_is'] = 'GO'
Monarch['tail_id_is'] = 'GO'
Monarch['relation'] = 'MolecularFunction_MolecularFunction'
Monarch['head_type'] = 'MolecularFunction'
Monarch['tail_type'] = 'MolecularFunction'
Monarch['kg_type'] = 'Generalised'
print(f"Monarch: {len(Monarch):,} rows")
Monarch
```

```python
# 2. PrimeKG (MolecularFunction -> MolecularFunction: Native)
PrimeKG = pd.read_csv(PROC_DIR + 'PrimeKG/PrimekG_MolFxn_MolFxn.csv')
PrimeKG.columns = PrimeKG.columns.str.lower()
required_columns = [
    'head', 'relation', 'tail', 'head_type', 'relation_type', 'tail_type',
    'kg_source', 'head_id_is', 'tail_id_is', 'head_detail_name', 'tail_detail_name'
]
PrimeKG = PrimeKG[required_columns]
PrimeKG['head_id_is'] = 'GO'
PrimeKG['tail_id_is'] = 'GO'
PrimeKG['relation'] = 'MolecularFunction_MolecularFunction'
PrimeKG['head_type'] = 'MolecularFunction'
PrimeKG['tail_type'] = 'MolecularFunction'
PrimeKG['kg_source'] = 'PrimeKG'
PrimeKG['kg_type'] = 'Generalised'
print(f"PrimeKG: {len(PrimeKG):,} rows")
PrimeKG
```

```python
# 3. GPKG (MolecularFunction -> MolecularFunction: Native)
GPKG = pd.read_csv(PROC_DIR + 'GPKG/MolecularFunction_MolecularFunction_GPKG.csv')
GPKG.columns = GPKG.columns.str.lower()
required_columns = [
    'head', 'relation', 'tail', 'head_type', 'relation_type', 'tail_type',
    'kg_source', 'head_id_is', 'tail_id_is', 'head_detail_name', 'tail_detail_name'
]
GPKG = GPKG[required_columns]
GPKG['head_id_is'] = 'GO'
GPKG['tail_id_is'] = 'GO'
GPKG['relation'] = 'MolecularFunction_MolecularFunction'
GPKG['head_type'] = 'MolecularFunction'
GPKG['tail_type'] = 'MolecularFunction'
GPKG['kg_source'] = 'GPKG'
GPKG['kg_type'] = 'Generalised'
print(f"GPKG: {len(GPKG):,} rows")
GPKG
```

```python
# 4. Crossbar (MolecularFunction -> MolecularFunction: Native)
Crossbar = pd.read_csv(PROC_DIR + 'crossbar/MolecularFunction_MolecularFunction_Crossbar.csv')
Crossbar.columns = Crossbar.columns.str.lower()
Crossbar['relation'] = 'MolecularFunction_MolecularFunction'
Crossbar['relation_type'] = 'NAN'
required_columns = [
    'head', 'relation', 'tail', 'head_type', 'relation_type', 'tail_type',
    'kg_source', 'head_id_is', 'tail_id_is', 'head_detail_name', 'tail_detail_name'
]
Crossbar = Crossbar[required_columns]
Crossbar['head_id_is'] = 'GO'
Crossbar['tail_id_is'] = 'GO'
Crossbar['head_type'] = 'MolecularFunction'
Crossbar['tail_type'] = 'MolecularFunction'
Crossbar['kg_source'] = 'Crossbar'
Crossbar['kg_type'] = 'Generalised'
print(f"Crossbar: {len(Crossbar):,} rows")
Crossbar
```

```python
# 5. TARKG (MolecularFunction -> MolecularFunction: Native)
TARKG1 = pd.read_csv(PROC_DIR + 'TARKG/MolecularFunction_MolecularFunction.csv')
TARKG1.columns = TARKG1.columns.str.lower()
required_columns = [
    'head', 'relation', 'tail', 'head_type', 'relation_type', 'tail_type',
    'kg_source', 'head_id_is', 'tail_id_is', 'head_detail_name', 'tail_detail_name'
]
TARKG1 = TARKG1[required_columns]
TARKG1['head_id_is'] = 'GO'
TARKG1['tail_id_is'] = 'GO'
TARKG1['relation'] = 'MolecularFunction_MolecularFunction'
TARKG1['head_type'] = 'MolecularFunction'
TARKG1['tail_type'] = 'MolecularFunction'
TARKG1['kg_source'] = 'TARKG'
TARKG1['kg_type'] = 'Generalised'
print(f"TARKG1: {len(TARKG1):,} rows")
TARKG1
```

```python
# 6. TARKG 2 (MolecularFunction -> MolecularFunction: Native)
TARKG2 = pd.read_csv(PROC_DIR + 'TARKG/MolecularFunction_MolecularFunction_2.csv')
TARKG2.columns = TARKG2.columns.str.lower()
required_columns = [
    'head', 'relation', 'tail', 'head_type', 'relation_type', 'tail_type',
    'kg_source', 'head_id_is', 'tail_id_is', 'head_detail_name', 'tail_detail_name'
]
TARKG2 = TARKG2[required_columns]
TARKG2['head_id_is'] = 'GO'
TARKG2['tail_id_is'] = 'GO'
TARKG2['relation'] = 'MolecularFunction_MolecularFunction'
TARKG2['head_type'] = 'MolecularFunction'
TARKG2['tail_type'] = 'MolecularFunction'
TARKG2['kg_source'] = 'TARKG'
TARKG2['kg_type'] = 'Generalised'
print(f"TARKG2: {len(TARKG2):,} rows")
TARKG2
```

```python
# PheKnowLator (MolecularFunction -> MolecularFunction: Native)
PHEKNOWLATOR = pd.read_csv(PROC_DIR + 'pheknowlator/MolecularFunction_MolecularFunction.csv')
PHEKNOWLATOR.columns = PHEKNOWLATOR.columns.str.lower()
PHEKNOWLATOR['relation_type'] = "NAN"
required_columns = [
    'head', 'relation', 'tail', 'head_type', 'relation_type', 'tail_type',
    'kg_source', 'head_id_is', 'tail_id_is', 'head_detail_name', 'tail_detail_name'
]
PHEKNOWLATOR = PHEKNOWLATOR[required_columns]

PHEKNOWLATOR['head_id_is'] = 'GO'
PHEKNOWLATOR['tail_id_is'] = 'GO'
PHEKNOWLATOR['relation'] = 'MolecularFunction_MolecularFunction'
PHEKNOWLATOR['head_type'] = 'MolecularFunction'
PHEKNOWLATOR['tail_type'] = 'MolecularFunction'
PHEKNOWLATOR['kg_source'] = 'PheKnowLator'
PHEKNOWLATOR['kg_type'] = 'Generalised'

print(f"PHEKNOWLATOR: {len(PHEKNOWLATOR):,} rows")
PHEKNOWLATOR
```

## 3 · Consolidate

```python
source_dfs = [Monarch, PrimeKG, GPKG, Crossbar, TARKG1, TARKG2, PHEKNOWLATOR]
aligned = []
for df in source_dfs:
    df = df.copy()
    df = df.loc[:, ~df.columns.duplicated()]
    for col in REQUIRED_COLS:
        if col not in df.columns:
            df[col] = None       
    aligned.append(df[REQUIRED_COLS])
final_df = pd.concat(aligned, ignore_index=True)
final_df = final_df[final_df['head'].astype(str).str.upper() != 'NAN']
final_df = final_df[final_df['tail'].astype(str).str.upper() != 'NAN']
final_df
```

## 4 · GO Name Normalisation & NA Audit Before Deduplication

```python
# Fill detail names using MESH GO dictionary (for both Head and Tail)
head_detail = final_df['head'].map(GO_dict)
final_df['head_detail_name'] = final_df['head_detail_name'].fillna(head_detail)

tail_detail = final_df['tail'].map(GO_dict)
final_df['tail_detail_name'] = final_df['tail_detail_name'].fillna(tail_detail)

final_df = final_df.dropna(subset=['head_detail_name', 'tail_detail_name']).reset_index(drop=True)

print(f"Consolidated rows (before deduplication): {len(final_df):,}")
print("\nNaN audit before deduplication:")
print(final_df.isna().sum())
```

## 5 · Deduplication & NA Audit After Deduplication

```python
def merge_sources(x):
    return '::'.join(sorted(set(x.dropna())))
group_cols = ['head', 'relation', 'tail']
final_df_dedup = final_df.groupby(group_cols, as_index=False).agg({
    'head_type': 'first', 'relation_type': 'first', 'tail_type': 'first',
    'kg_source': merge_sources, 'kg_type': merge_sources,
    'head_id_is': 'first', 'tail_id_is': 'first',
    'head_detail_name': 'first', 'tail_detail_name': 'first'
})

print(f"Before dedup: {len(final_df):,}  |  After dedup: {len(final_df_dedup):,}")
print("\nNaN audit after deduplication:")
print(final_df_dedup.isna().sum())
```

## 6 · Save Output

```python
os.makedirs(os.path.dirname(OUT_PATH), exist_ok=True)
final_df_dedup.to_csv(OUT_PATH, index=False)
print(f"Saved {len(final_df_dedup):,} rows → {OUT_PATH}")
```
