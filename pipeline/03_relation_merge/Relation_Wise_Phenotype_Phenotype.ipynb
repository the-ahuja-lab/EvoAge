# Phenotype ↔ Phenotype Relation-Wise Merge

Merges Phenotype–Phenotype triples from PrimeKG, GPKG, and TARKG;
deduplicates by `(head, relation, tail)`; and saves the result.

## 0. Configuration

```python

import os
import pandas as pd
import numpy as np

BASE_DIR  = '/storage/Arushi/090526_EvoAge/kg_formation/'
DB_DIR     = BASE_DIR + 'data_collection/databases_for_mapping/'
PROC_DIR  = BASE_DIR + 'processed_data/'

# ── Output path ───────────────────────────────────────────────────────────────
OUT_PATH   = BASE_DIR + 'processed_data_relation_wise_merge/generalised/PHENOTYPE_PHENOTYPE/ALL_PHENOTYPE_PHENOTYPE.csv'

# ── Required output schema ────────────────────────────────────────────────────

REQUIRED_COLS = [
    'head', 'relation', 'tail',
    'head_type', 'relation_type', 'tail_type',
    'kg_source',
    'head_id_is', 'tail_id_is',
    'head_detail_name', 'tail_detail_name', 'kg_type', 'species'
]
```

## 1. Load KG Sources

### 1a. PrimeKG

```python
PrimeKG_Phenotype_Phenotype = pd.read_csv(PROC_DIR + 'PrimeKG/PrimeKG_Phenotype_Phenotype.csv')
PrimeKG_Phenotype_Phenotype.columns = PrimeKG_Phenotype_Phenotype.columns.str.lower()
print(f"PrimeKG: {len(PrimeKG_Phenotype_Phenotype):,} rows | columns: {list(PrimeKG_Phenotype_Phenotype.columns)}")
PrimeKG_Phenotype_Phenotype['kg_type'] = 'Generalised'
PrimeKG_Phenotype_Phenotype['kg_source'] = 'PrimeKG'
PrimeKG_Phenotype_Phenotype['species'] = 'Homo species'
PrimeKG_Phenotype_Phenotype.head(2)
```

### 1b. GPKG

```python
GPKG_Phenotype_Phenotype = pd.read_csv(PROC_DIR + 'GPKG/Phenotype_Phenotype_GPKG.csv')
GPKG_Phenotype_Phenotype.columns = GPKG_Phenotype_Phenotype.columns.str.lower()
GPKG_Phenotype_Phenotype.rename(columns={'head_name': 'head_detail_name', 'tail_name': 'tail_detail_name'}, inplace=True)
print(f"GPKG:    {len(GPKG_Phenotype_Phenotype):,} rows | columns: {list(GPKG_Phenotype_Phenotype.columns)}")
GPKG_Phenotype_Phenotype['kg_type'] = 'Generalised'
GPKG_Phenotype_Phenotype['kg_source'] = 'GPKG'
GPKG_Phenotype_Phenotype['species'] = 'Homo species'

GPKG_Phenotype_Phenotype.head(2)
```

### 1c. TARKG

```python
TARKG_Phenotype_Phenotype = pd.read_csv(PROC_DIR + 'TARKG/Phenotype_Phenotype.csv')
TARKG_Phenotype_Phenotype.columns = TARKG_Phenotype_Phenotype.columns.str.lower()
print(f"TARKG:   {len(TARKG_Phenotype_Phenotype):,} rows | columns: {list(TARKG_Phenotype_Phenotype.columns)}")
GPKG_Phenotype_Phenotype['kg_type'] = 'Generalised'
GPKG_Phenotype_Phenotype['kg_source'] = 'TARKG'
GPKG_Phenotype_Phenotype['species'] = 'Homo species'
TARKG_Phenotype_Phenotype.head(2)
```

### 1d. pheknowlator

```python
pheknowlator_Phenotype_Phenotype = pd.read_csv(PROC_DIR + 'pheknowlator/Phenotype_Phenotype.csv')
pheknowlator_Phenotype_Phenotype.columns = pheknowlator_Phenotype_Phenotype.columns.str.lower()
print(f"pheknowlator:   {len(pheknowlator_Phenotype_Phenotype):,} rows | columns: {list(pheknowlator_Phenotype_Phenotype.columns)}")
pheknowlator_Phenotype_Phenotype['kg_type'] = 'Generalised'
pheknowlator_Phenotype_Phenotype['kg_source'] = 'pheknowlator'
pheknowlator_Phenotype_Phenotype['species'] = 'Homo species'
pheknowlator_Phenotype_Phenotype.head(2)
```

## 2. Check and Fix Duplicate Columns

```python
SOURCE_DFS = [
    ('PrimeKG_Phenotype_Phenotype', PrimeKG_Phenotype_Phenotype),
    ('GPKG_Phenotype_Phenotype',    GPKG_Phenotype_Phenotype),
    ('TARKG_Phenotype_Phenotype',   TARKG_Phenotype_Phenotype),
    ('pheknowlator_Phenotype_Phenotype', pheknowlator_Phenotype_Phenotype),
]

for name, df in SOURCE_DFS:
    dupes = df.columns[df.columns.duplicated(keep=False)].unique().tolist()
    if dupes:
        print(f"\n[{name}] duplicate columns: {dupes}")
        for col in dupes:
            for i, (_, s) in enumerate(df.loc[:, df.columns == col].items()):
                print(f"  '{col}' occurrence {i} → sample: {s.dropna().head(3).tolist()}")
    else:
        print(f"[{name}] ✓ no duplicates")
```

## 3. Align Schemas and Concatenate

```python
CONCAT_COLS = [
    'head', 'relation', 'tail',
    'head_type', 'relation_type', 'tail_type',
    'kg_source', 'head_id_is', 'tail_id_is',
    'head_detail_name', 'tail_detail_name', 'kg_type', 'species'
]

df_list = []
for name, df in SOURCE_DFS:
    tmp = df.copy()
    for col in CONCAT_COLS:
        if col not in tmp.columns:
            tmp[col] = pd.NA
    df_list.append(tmp[CONCAT_COLS])

final_df = pd.concat(df_list, ignore_index=True)
print(f"Combined: {len(final_df):,} rows")
final_df.head(2)
```

## 4. Standardise Fixed-Value Columns

```python
final_df['head_type'] = 'Phenotype'
final_df['tail_type'] = 'Phenotype'
final_df['relation']  = 'Phenotype_Phenotype'
final_df['head_id_is'] = 'HPO'
final_df['tail_id_is'] = 'HPO'

print("Unique relation:",   set(final_df['relation']))
print("Unique head_type:",  set(final_df['head_type']))
print("Unique tail_type:",  set(final_df['tail_type']))
print("Unique kg_source:",  set(final_df['kg_source']))
```

## 5. Deduplicate and Add Schema Columns

Group by `(head, relation, tail)` and collapse `kg_source` with `::` separator.

```python
GROUP_COLS = ['head', 'relation', 'tail']

def merge_sources(x):
    return '::'.join(sorted(set(x.dropna())))

print(f"Before deduplication: {len(final_df):,} rows")
final_df_dedup = final_df.groupby(GROUP_COLS, as_index=False).agg({
    'head_type':        'first',
    'relation_type':    'first',
    'tail_type':        'first',
    'kg_source':         merge_sources,
    'head_id_is':       'first',
    'tail_id_is':       'first',
    'head_detail_name': 'first',
    'tail_detail_name': 'first',
    'kg_type': merge_sources,
    'species': 'first',
    
})


final_df_dedup = final_df_dedup[REQUIRED_COLS]

print(f"After deduplication: {len(final_df_dedup):,} rows")
final_df_dedup.head()
```

## 6. QC — NaN Check

```python
def nan_summary(df):
    true_nan   = df.isna().sum()
    string_nan = df.apply(lambda c: c.astype(str).str.upper().eq('NAN').sum())
    return pd.DataFrame({
        'NaN_count':          true_nan,
        "'NAN'_string_count": string_nan,
        'Total_NaN_like':     true_nan + string_nan
    })

nan_summary(final_df_dedup)
```

## 7. Save Output

```python
final_df_dedup.to_csv(OUT_PATH, index=False)
print(f"Saved {len(final_df_dedup):,} rows → {OUT_PATH}")
```
