```python
import os
import re
import pandas as pd

# Path to your log files
log_dir = "../3-TrainedModel_prediction/Edge_type_Metrics/"

# Regex to match metrics
metrics_pattern = re.compile(r"(MRR|Hits@1|Hits@3|Hits@10)\s*=\s*([\d.]+)")

data = []

for filename in os.listdir(log_dir):
    if filename.endswith(".log"):
        filepath = os.path.join(log_dir, filename)
        
        # Read file content
        with open(filepath, "r") as f:
            text = f.read()
        
        # Extract all metrics from file
        metrics = dict(metrics_pattern.findall(text))
        
        # Convert values to float
        metrics = {k: float(v) for k, v in metrics.items()}
        
        # Add filename for reference
        metrics["file"] = filename
        
        data.append(metrics)

# Create DataFrame
df = pd.DataFrame(data)

# Reorder columns
df = df[["file", "MRR", "Hits@1", "Hits@3", "Hits@10"]]


# Save to CSV
df.to_csv("edge_metrics_summary.csv", index=False)
df
```
