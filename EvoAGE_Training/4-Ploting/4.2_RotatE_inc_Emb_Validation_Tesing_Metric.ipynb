```python
import os
import re
import pandas as pd
from collections import defaultdict

# Set path to your log files directory
log_dir = "../2-DGL_Training/Training_files/"

# Target log files and their readable names
target_logs = {
    "rotatE_64_emb_final.log": "RotatE_64_Emb",
    "rotatE_128_emb_128Hidim_final.log": "RotatE_128_Emb",
    "rotatE_256_emb_256Hidim_final.log": "RotatE_256_Emb",
    "rotatE_512_emb_512Hidim_final.log": "RotatE_512_Emb",
    
}

# Dictionary to collect metric values (lists for averaging duplicates)
metrics_all = defaultdict(lambda: defaultdict(list))

# Loop through selected logs only
for filename, model_name in target_logs.items():
    filepath = os.path.join(log_dir, filename)
    with open(filepath, 'r') as file:
        for line in file:
            if "Valid average" in line:
                match = re.search(r"Valid average (\w+@?\d*): ([\d.]+)", line)
                if match:
                    metric, value = match.groups()
                    metric = metric.upper()
                    metrics_all[model_name][metric].append(float(value))

# Average all collected values per model & metric
metrics_avg = {
    model: {
        metric: sum(values) / len(values)
        for metric, values in metric_values.items()
    }
    for model, metric_values in metrics_all.items()
}

# Convert to DataFrame
df = pd.DataFrame.from_dict(metrics_avg, orient="index")

# Rename & reorder
df.rename(columns={
    "HITS@1": "Hit@1", "HITS@3": "Hit@3", "HITS@10": "Hit@10"
}, inplace=True)

# Ensure expected column order (ignore missing gracefully)
ordered_cols = [col for col in ["Hit@1", "Hit@3", "Hit@10", "MR", "MRR"] if col in df.columns]
df = df[ordered_cols]

# Add index as "Model Type"
df.reset_index(inplace=True)
df.rename(columns={"index": "Model Type"}, inplace=True)
df.to_csv('RotatE_Increasing_embedding_Validation_Results.csv',index = None)
df
```
