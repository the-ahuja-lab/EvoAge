```python
# Global mapping to ensure consistency across CSV files
global_mapping = {}
```

```python
from functools import reduce

def create_mapping_of_nodes_from_pickle(path ,header_list):
    global global_mapping

    if(path.lower().endswith(('.csv'))):
        df = pd.read_csv(path, compression='infer')
    else:
        df = pd.read_pickle(path, compression='infer')

    uniques_per_header = []

    for header in header_list:
        unique_head = df[header].astype(str).unique()
        uniques_per_header.append(unique_head)

    unique_nodes = reduce(np.union1d,uniques_per_header)
    print('Total Unique In This: ',len(unique_nodes))

    for name in unique_nodes:
        if name not in global_mapping:
            global_mapping[name] = len(global_mapping)

    print("Final Mapping Size: ",len(global_mapping))
```

# Function to convert triples in MappedID format

```python

def triples_as_mappedID(path, head_type, tail_type):
    if path.lower().endswith('.csv'):
        df = pd.read_csv(path, compression='infer')
    else:
        df = pd.read_pickle(path, compression='infer')

    head_map_path = f'Store_House/{head_type}_id_H_KG.pkl'
    tail_map_path = f'Store_House/{tail_type}_id_H_KG.pkl'

    head_df = pd.read_pickle(head_map_path, compression='infer')
    tail_df = pd.read_pickle(tail_map_path, compression='infer')

    head_type_cap = head_type.capitalize()
    tail_type_cap = tail_type.capitalize()

    # Ensure consistent dtype (string) for merge columns
    df['head'] = df['head'].astype(str)
    df['tail'] = df['tail'].astype(str)
    head_df[head_type_cap] = head_df[head_type_cap].astype(str)
    tail_df[tail_type_cap] = tail_df[tail_type_cap].astype(str)

    # Merge and map head IDs
    head_id = pd.merge(df[['head']], head_df,
                       left_on='head', right_on=head_type_cap, how='left')
    head_id_tensor = torch.from_numpy(head_id['MappedID'].values).to(dtype=torch.int64)
    print('Head Tensor:', head_id_tensor)

    # Merge and map tail IDs
    tail_id = pd.merge(df[['tail']], tail_df,
                       left_on='tail', right_on=tail_type_cap, how='left')
    tail_id_tensor = torch.from_numpy(tail_id['MappedID'].values).to(dtype=torch.int64)
    print('Tail Tensor:', tail_id_tensor)

    # Build COO edge index
    edge_index = torch.stack([head_id_tensor, tail_id_tensor], dim=0)
    print('\nEdges in COO format:')
    print(edge_index)

    return edge_index
```
